```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=pd.read_csv("/content/World Indicators.csv")
dt.head()
```

```python
dt.dropna(inplace = True)
```

```python
dt.drop_duplicates(inplace = True)
dt.head()
```

```python
dt.dtypes
```

```python
#Remove special characters like '$' and '%' from individual records and convert string to float.
dt['GDP'] = dt['GDP'].replace('[\$,]', '', regex=True).astype(float)
dt['Health Exp/Capita'] = dt['Health Exp/Capita'].replace('[\$,]', '', regex=True).astype(float)
dt['Business Tax Rate'] = pd.to_numeric(dt['Business Tax Rate'].replace({'%':''} , regex=True))

dt.dtypes
```

```python
dt.head()
```

```python
#selecting numerical data and standardization of varaibles
dt_x = dt.iloc[:, :18]
dt_y=dt[['Region','Country']]
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
for i in dt_x.columns:
  dt_x[[i]] = scaler.fit_transform(dt_x[[i]])
dt_x.head()
```

```python
#kmeans clustering
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt  
#elbow method
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i,init='k-means++', random_state=42)
    k_means.fit(dt_x)
    wcss.append(k_means.inertia_)
#plot elbow curve
plt.plot(np.arange(1,11),wcss)
plt.xlabel('Clusters')
plt.ylabel('sum of squared distance(wcss)')
plt.show()
```

```python
# Perform K-means clustering with k=2
kmeans = KMeans(n_clusters=2).fit(dt_x)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
# Calculate Silhoutte Score
score = silhouette_score(dt_x,kmeans.labels_, metric='euclidean')
# Print the score
print('Silhouetter Score: %.3f' % score)
```

```python
# Perform hierarchical clustering
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(dt_x, 'ward')

# Plot the dendrogram
plt.figure()
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('samples')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity = "euclidean", linkage = "ward")
cluster = hc.fit_predict(dt_x)
```

```python
# Calculate Silhoutte Score
score = silhouette_score(dt_x,hc.labels_, metric='euclidean')
# Print the score
print('Silhouetter Score: %.3f' % score)
```

```python
#Mapping Country and Region back
dt_x['kmeans_result']=pd.Series(kmeans.labels_)
dt_x.rename(columns={"Region": "Region_T", "Country": "Country_T"}, inplace=True)
dt_x = pd.concat([dt_x,dt_y], axis=1)

dt_x.head()
```

```python
cluster_1_countries= dt_x[dt_x['kmeans_result'] == 0] 
cluster_2_countries= dt_x[dt_x['kmeans_result'] == 1]
cluster_1_countries['Country'].unique()
cluster_2_countries['Country'].unique()
```
